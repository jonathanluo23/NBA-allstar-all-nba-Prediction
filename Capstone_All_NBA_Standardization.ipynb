#Preprocessing 

##Skew

```python
continous.skew()
```

```python
from scipy import stats as stats
def yeo(df):
  new_df = []
  for col in df.columns:
    # print(col)
    # print(df.loc[:,col])
    new_col = stats.yeojohnson(df.loc[:,col])[0]
    # print(new_col.shape)
    new_df.append(new_col)
  new_df = np.transpose(np.array(new_df))
  new_df = pd.DataFrame(new_df, columns=df.columns)
  return(new_df)
```

```python
new_continuous = yeo(continous)
```

```python
new_continuous.skew()
```

##Scale

```python
from sklearn.preprocessing import StandardScaler
def standard(data):
  standardscaler = StandardScaler()
  datanorm= standardscaler.fit_transform(data)
  ind=data.index
  cols =data.columns
  datanorm = pd.DataFrame(datanorm, index=ind, columns= cols)
  return (datanorm)
```

```python
newcont_scaled = standard(new_continuous)
newcont_scaled.head()
```
